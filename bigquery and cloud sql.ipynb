```python
from google.cloud import bigquery
client = bigquery.Client()
```

```python
sql = """
  SELECT
    fog,
    COUNT(fog) AS num_fog,
    AVG(mean_visibility) as mean_visibility
  FROM
    `bigquery-public-data.samples.gsod`
  WHERE
    year >= 2000 AND year <= 2010	And NOT IS_NAN(mean_visibility)
  GROUP BY
    fog
"""
df = client.query(sql).to_dataframe()
df.head()
```

```python
sql = """
  SELECT 
    fog,
    COUNT(fog) AS num_fog,
    year
  FROM
    `bigquery-public-data.samples.gsod`
  WHERE
    year >= 2000 AND year <= 2010
  GROUP BY
    fog, year
  ORDER BY
    year DESC, fog DESC
"""
df = client.query(sql).to_dataframe()
df.head()
```

```python
pivot_table = df.pivot(index='year', columns='fog', values='num_fog')
pivot_table.plot(kind='bar', stacked=True);
```

```python
sql = """
  SELECT
    fog,
    AVG(mean_temp) as mean_temp
  FROM
    `bigquery-public-data.samples.gsod`
  WHERE
    year >= 2000 AND year <= 2010
  GROUP BY
    fog
"""
df = client.query(sql).to_dataframe()
df.plot(x='fog', y='mean_temp', kind='bar');
```

```python
sql = """
  SELECT
    month,
    AVG(max_sustained_wind_speed) as max_sustained_wind_speed
  FROM
    `bigquery-public-data.samples.gsod`
  WHERE
    year >= 2000 AND year <= 2010
  GROUP BY
    month
  ORDER BY
    month
"""
df = client.query(sql).to_dataframe()
df.plot(x='month', y='max_sustained_wind_speed', kind='bar');
```

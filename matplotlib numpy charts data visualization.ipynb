# 1. Adolescents Sports Group Bar Plot

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

#First set of data
male = (40, 24, 20, 17, 9, 8, 5)
ind = np.arange(7)  # the x locations for the groups
width = 0.25       # the width of the bars
fig, ax = plt.subplots()

#Creating the first set of bars for the first set of data
rects1 = ax.bar(ind, male, width, color='blue') 

#Second set of data-- creation of bars 
female = (25, 17, 17, 17, 12, 8, 4)
rects2 = ax.bar(ind + width, female, width, color='red')

# Add some text for labels, title and axes ticks
ax.set_ylabel('Percentage (%) of Adolescents',fontsize=13)
plt.xlabel('Sports',fontsize=10,labelpad=15)
ax.set_title('Adolescents Sports',fontsize=10)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Basket', 'Base/Soft ball','Soccer', 'Track','Swimming','Tennis','Cross Country'),fontsize=8)

#Create a legend 
ax.legend((rects1[0], rects2[0]), ('Male', 'Female'))

#save into a raster graphics format
plt.savefig('adolescents_sports.png')
plt.show()
```


# 4. 3D Plot (Gabor Function)


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
#Following line is for color mapping
from matplotlib import cm

#Creates an unfilled contour with a cool & warm color mapping

#Creating the data for the contour plot(s)
delta = 0.025
x = np.arange(0,9,delta)
y = np.arange(0,9,delta)
#meshgrid makes rectangular arrays that
#cover every combination of x and y values 
X, Y = np.meshgrid(x,y)

#make an interesting bump and valley for plotting
#Creates a Gaussian distribution for equal shape X, Y
# Z1 = mlab.bivariate_normal(X,Y,1.0,1.0,0.0,0.0)
# Z2 = mlab.bivariate_normal(X,Y,1.5,0.5,1,1)
Z = np.exp(-0.5 * (X**2/4 + Y**2/4)) * np.sin(2 * np.pi * 0.2 * X)

#Creating an unfilled contour plot
plt.figure()
#Can change the color mappings and line style
cp = plt.contour(X, Y, Z,cmap=cm.coolwarm)

#Labels and titles for the plot
plt.clabel(cp, inline=True, fontsize=10)
plt.title('Cell receptive field in primary visual cortex',fontsize=14)
plt.xlabel('Optimal brightness of the image (cm)')
plt.ylabel('Visual neuron response (cm)')
plt.show()
```

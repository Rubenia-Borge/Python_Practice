```python
# Number 1
var = "Hello, World!"
print ("Number 1:", var)

# Number 2
print ("Number 2: The length of variable var is",len(var))

# Number 3
print ("Number 3: ",var[7:12])

# Number 4
print ("Number 4: ",type(1), type(1.0), type("1"))

# Number 5
import math
print ("Number 5: The area of a circle with radio = 5 is ",round((pow(5,2)*math.pi),2))

# Number 6
print ("Number 6: For Loop numbers 0 to 9")
for i in range(0,10):
  print(i)
print(" ")

# Number 7
print ("Number 7: For Loop even number 10 to 20 ")
for i in range(10,21):
    if (i % 2 == 0):
        print(i)
print(" ")

# Number 8
print("Number 8: If statement")
if len(var) > 10:
    print("Greater than 10")
elif len(var) < 10:
    print("Less than 10")
elif len(var) == 10:
    print("Equal to 10")
else:
    print("N/A")

# Number 9
print("\nNumber 9: Function String")
# Defining the Function
def myStringFunction(myStrParam):
  my_length = len(myStrParam)
  print("This function returns the length of the String that it received as a parameter: ")
  return(my_length)

# Calling the function
myStrParam = "This is the text passed to the function!"
print("This is the length: ", myStringFunction(myStrParam))

# Number 10
print("\nNumber 10: Function Integer")
# Defining the Function
def myIntegerFunction(myIntParam):
    for i in range(1, myIntParam+1):
      if (i % 2 != 0):
        print(i)
    print(" ")
    return 

# Calling the function
myIntParam = 5
print("This function receives an integer n as a parameter and returns the odd numbers from 1 to n: ")
myIntegerFunction(myIntParam)
    

```
